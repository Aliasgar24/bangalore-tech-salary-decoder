# Task 2 - Data Cleaning

```python
# Rename the columns, this will make the columns in same format, like all lowercase letters

df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()

print(df.columns)
```

```python
# Now check for unique roles

df['role'].unique()
```

```python
# assign diffrent roles a comman role name

role_mapping = {

    # Data Scientist
    "DS": "Data Scientist",

    # DevOps
    "DevOps": "DevOps Engineer",
    "DevOps Engineer": "DevOps Engineer",

    # Backend
    "Backend Dev": "SDE Backend",
    "Backend Developer": "SDE Backend",
    "Backend Engineer": "SDE Backend",
    "SDE-Backend": "SDE Backend",
    "BE": "SDE Backend",

    # Frontend
    "Frontend Dev": "SDE Frontend",
    "Frontend Engineer": "SDE Frontend",
    "FE": "SDE Frontend",
    "SDE-Frontend": "SDE Frontend",
    "Frontend Developer": "SDE Frontend",

    # Full Stack
    "Full Stack Engineer": "SDE Full-Stack",
    "FullStack": "SDE Full-Stack",
    "Fullstack Dev": "SDE Full-Stack",

    # Product
    "Product Lead": "Product Manager",
    "PM": "Product Manager",

    # UI/UX
    "UI Designer": "UI/UX Designer",
    "Designer": "UI/UX Designer",
    "UX Designer": "UI/UX Designer",
    "UI/UX": "UI/UX Designer",

    # Data Analyst
    "DA": "Data Analyst",

    # Business Analyst
    "BA": "Business Analyst",
    "BI Analyst": "Business Analyst",
    "Business Systems Analyst": "Business Analyst",

    # Site Reliability Engineer
    "SRE": "Site Reliability Engineer",

    # Others
    "ML Engineer": "ML Engineer",
    "Data Science Engineer": "Data Science Engineer",
    "Analytics Engineer": "Analytics Engineer",
    "Infra Engineer": "Infra Engineer",
    "SDE FS": "SDE Full-Stack",
    "Sr PM": "Senior Product Manager"
}

df["role_clean"] = df["role"].replace(role_mapping)

df["role_clean"].value_counts()
```

```python
# Lets check unique values for company type

df["company_type"].unique()
```

```python
# Lets group together the similar ones

company_mapping = {
    "Unicorn": "Unicorn",
    "unicorn": "Unicorn",
    "UNICORN": "Unicorn",

    "MNC": "MNC",
    "mnc": "MNC",

    "Mid-size": "Mid-size",
    "mid-size": "Mid-size",
    "MID-SIZE": "Mid-size",

    "Early-stage": "Early-stage",
    "early-stage": "Early-stage",
    "EARLY-STAGE": "Early-stage"
}

df["company_type_clean"] = df["company_type"].replace(company_mapping)

df["company_type_clean"].value_counts()
```

```python
# Now for education Tier

df["education_tier"].unique()
```

```python
# Lets group them together

education_mapping = {
    "Tier-1": "Tier 1",
    "Tier 1": "Tier 1",
    "T1": "Tier 1",
    "1": "Tier 1",

    "Tier-2": "Tier 2",
    "Tier 2": "Tier 2",
    "T2": "Tier 2",
    "2": "Tier 2",

    "Tier-3": "Tier 3",
    "Tier 3": "Tier 3",
    "T3": "Tier 3",
    "3": "Tier 3"
}

df["education_tier_clean"] = df["education_tier"].replace(education_mapping)

df["education_tier_clean"].value_counts()
```

```python
# Now we check the diff formats used in current_ctc

df["current_ctc"].head(20)
```

```python
# Now we need to clean the current_ctc and bring it into a single float format (LPA)



def parse_ctc(value):

    if pd.isna(value):
        return np.nan

    value = str(value)

    # Now we Remove unnecessary symbols like rupee, LPA, the commas
    value = value.replace("₹", "")
    value = value.replace("LPA", "")
    value = value.replace(",", "")
    value = value.strip()

    salary = float(value)

    # Convert rupees into LPA
    if salary > 100000:
        salary = salary / 100000

    return salary


df["current_ctc_lpa"] = df["current_ctc"].apply(parse_ctc)

print(df["current_ctc_lpa"].head(20))
```

```python
# As we see, all salaries are converted to a single format
```

```python
# Now we Convert Previous_CTC into float (LPA), we use the functoni created above def parse_ctc and pass the previous_ctc values

df["previous_ctc_lpa"] = df["previous_ctc"].apply(parse_ctc)

print(df["previous_ctc_lpa"].head(20))
```

```python
# Fill missing values in Skills and Location

df["skills"] = df["skills"].fillna("Unknown")
df["location"] = df["location"].fillna("Unknown")
```

```python
# Remove duplicate rows from the dataset

df = df.drop_duplicates()

print("Number of rows after removing duplicates:", len(df))
```

# Task 3.1 - CTC Distribution by Role

```python
# Calculate salary statistics for each role

role_salary = df.groupby("role_clean")["current_ctc_lpa"].agg(
    ["median", "mean", "min", "max"]
)

# Sort roles by median salary in descending order

role_salary = role_salary.sort_values(by="median", ascending=False)

print(role_salary)
```

```python
# highest and lowest paying roles based on median salary

highest_role = role_salary.index[0]
lowest_role = role_salary.index[-1]

print("Highest Paying Role:", highest_role)
print("Lowest Paying Role:", lowest_role)
```

```python
# Compare mean and median salary for each role

print("Mean vs Median Salary by Role")
print(role_salary[["median", "mean"]])
```



Senior Product Managers have the highest median salary (31.5 LPA), while Data Analysts have the lowest (16.3 LPA).

# Task 3.2 - Experience Curve for SDE Backend

```python
# Select only SDE Backend employees

backend_df = df[df["role_clean"] == "SDE Backend"].copy()

print("Total SDE Backend Employees:", len(backend_df))
```

```python
# Divide employees into experience groups

bins = [-1, 1, 3, 5, 100]
labels = ["0-1", "2-3", "4-5", "6+"]

backend_df["exp_band"] = pd.cut(
    backend_df["years_exp"],
    bins=bins,
    labels=labels
)

print(backend_df[["years_exp", "exp_band"]].head())
```

```python
# Calculate median salary for each experience band

experience_salary = backend_df.groupby("exp_band")["current_ctc_lpa"].median()

print(experience_salary)
```

```python
# Calculate salary growth for each experience band

growth_1 = ((experience_salary["2-3"] - experience_salary["0-1"]) / experience_salary["0-1"]) * 100

growth_2 = ((experience_salary["4-5"] - experience_salary["2-3"]) / experience_salary["2-3"]) * 100

growth_3 = ((experience_salary["6+"] - experience_salary["4-5"]) / experience_salary["4-5"]) * 100

print("Growth from 0-1 to 2-3 years :", round(growth_1, 2), "%")
print("Growth from 2-3 to 4-5 years :", round(growth_2, 2), "%")
print("Growth from 4-5 to 6+ years  :", round(growth_3, 2), "%")
```



The median salary of SDE Backend professionals increases steadily with experience, from 11.65 LPA for 0–1 years to 40.40 LPA for professionals with more than 6 years of experience.

# Task 3.3 - Skill Premium for SDEs

```python
# Select all SDE roles

sde_df = df[
    (df["role_clean"] == "SDE Backend") |
    (df["role_clean"] == "SDE Frontend") |
    (df["role_clean"] == "SDE Full-Stack")
].copy()

print("Total SDE Employees:", len(sde_df))
```

```python
# List of skills to analyse, we will select these 4 skills

skills = ["AWS", "ML", "System Design", "Kubernetes"]
```

```python
# Compare median salary with and without each skill

for skill in skills:

    with_skill = sde_df[
        sde_df["skills"].str.contains(skill, case=False, na=False)
    ]["current_ctc_lpa"].median()

    without_skill = sde_df[
        ~sde_df["skills"].str.contains(skill, case=False, na=False)
    ]["current_ctc_lpa"].median()

    premium = ((with_skill - without_skill) / without_skill) * 100

    print("Skill :", skill)
    print("Median with skill :", round(with_skill, 2), "LPA")
    print("Median without skill :", round(without_skill, 2), "LPA")
    print("Premium :", round(premium, 2), "%")
    print("-" * 40)
```

among the skills we chose, System Design provides highest salary premium (20.48%), whereas AWS provides a negative premium of (-4.34%)

```python
# Calculate median salary for each company type

company_salary = backend_df.groupby("company_type_clean")["current_ctc_lpa"].median()

print(company_salary)
```

# Task 3.5 - Underpaid Professionals

```python
# Create experience bands for the entire dataset

bins = [-1, 1, 3, 5, 100]
labels = ["0-1", "2-3", "4-5", "6+"]

df["exp_band"] = pd.cut(
    df["years_exp"],
    bins=bins,
    labels=labels
)

print(df[["years_exp", "exp_band"]].head())
```

```python
# Keep  groups that have at least 10 employees

valid_groups = df.groupby(
    ["role_clean", "company_type_clean", "exp_band"]
).filter(lambda x: len(x) >= 10)

print("Valid Rows:", len(valid_groups))
```

```python
# Calculate the median salary of each group

valid_groups["group_median"] = valid_groups.groupby(
    ["role_clean", "company_type_clean", "exp_band"]
)["current_ctc_lpa"].transform("median")
```

```python
# Calculate salary gap from the group median

valid_groups["gap"] = (
    valid_groups["current_ctc_lpa"] -
    valid_groups["group_median"]
)
```

```python
# Display the top 10 most underpaid professionals

underpaid = valid_groups.sort_values(by="gap").head(10)

print(
    underpaid[
        [
            "employee_id",
            "role_clean",
            "company_type_clean",
            "years_exp",
            "current_ctc_lpa",
            "group_median",
            "gap"
        ]
    ]
)
```



The most underpaid employee in the filtered groups is BLR0954 (Business Analyst), earning 25.40 LPA, which is 12.25 LPA below the median salary of their peer group.
